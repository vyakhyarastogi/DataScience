```python
#question1

class vehicle:
    def __init__(self,name_of_vehicle,max_speed,average_of_vehicle):
        self.name=name_of_vehicle
        self.speed=max_speed
        self.average=average_of_vehicle
        
    def return_details(self):
        return self.name,self.speed,self.average
    
transport=vehicle("Mercedes-Benz",500,300)
transport.return_details()


#QUESTION-2

class car(vehicle):
    def seating_capacity(self,capacity):
        self.capacity=capacity
        return self.name,self.capacity

obj_car=car("benz",20,30)
obj_car.seating_capacity(10)
```

```python
#Question3

''' Multiple inheritance means one class will inherit the properties of other classess like class3 inherit the property of class1 and class2'''

class class1:
    def test_class1(self):
        return "this is a method inside class1"

class class2:
    def test_class2(self):
        return "this is a method inside class2"

class class3(class1,class2):
    pass

obj_class3=class3()
obj_class3.test_class1()

        

```

```python
#QUESTION-4

'''getter and setter are used to access the data which has been inside the private'''

class info:
    def __init__(self,name,salary,age):
        self.name=name
        self.__salary=salary # private
        self.age=age
        
     #setter   
    def set_salary(self,salary):
        self.__salary=salary
    
    #getter
    def get_salary(self):
        return self.__salary
        
information=info("chota",200,20)
information.set_salary(5000)
information.get_salary()
```

```python
#QUESTION-5
'''Method overriding is an ability of any object-oriented programming language that allows a subclass or child class to provide a specific implementation of a method that is already provided by one of its super-classes or parent classes.'''

class parent:
    def __init__(self):
        self.value="inside parent"
    
    
    #parent's show method
    
    def show(self):
        print(self.value)
        

#defining a child class
class child(parent):
    def __init__(self):
        self.value="inside child"
        
    #child's show method
    
    def show(self):
        print(self.value)
        
obj1=parent()
obj2=child()

obj1.show()
obj2.show()
        
```
