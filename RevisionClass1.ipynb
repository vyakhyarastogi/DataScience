```python
#### we have to build a student database wherein we will have:
          *Student_ids
          *Gender
          *Department
          *Sessional_perc
          *Final_sem marks
          *Backlogs
          
          
          
          
          
       
```

```python
student_ids=['S01','S02','S03','S04','S05']
gender=['M','F','F','M','F']
dept=['CS','ECE','ME','CS','ME']
sessional_perc=[45,60,73,65.3,71.9]
total_marks=[680,800,520,650,780]
backlogs=['NB','B','NB','B','NB']


```

 
if any student has recieved more than 70% in their sessional increase the total marks by 2%

```python
# for i in sessional_perc:
#    if(i>70):

for i in range(len(student_ids)):
    if sessional_perc[i]>70:
        total_marks[i]=total_marks[i]+total_marks[i]*0.02
    else:
        print("marks did not change")
    
        
```

#### S01 and S04  have got updated backlog notice.S01 has got backlog and S04 backlog has been cleared

```python
# since lists are mutable data type
for i in range(len(student_ids)):
    if(student_ids[i]=='S01'):
        backlogs[i]='B'
    elif (student_ids[i]=='S04'):
        backlogs[i]='NB'
    else:
        print("Do nothing,take a chill, all ok")
```

#### If a students total marks is greater than 750 and they have a no backlog they will receive a 'high achiever scholarship'

If a student has no backlog and has marks between 650 and 700 they will receive 'Achiever Scholarship'

```python
scholarships=[]
for i in range(len(student_ids)):
    scholarships.append('No Scholarhsips')
for i in range(len(student_ids)):
    if (total_marks[i]>750 and backlogs[i]=='NB'):
        scholarships[i]='High Achiever Scholarship'
        print(f"student {student_ids[i]} receive a 'high acheiver scholarship'")
    elif (total_marks[i]>=650 and total_marks[i]<=700 and backlogs[i]=='NB'):
        scholarships[i]='Achiever Scholarship'
        print(f"student {student_ids[i]} to receive a 'achiver scholarship'")
    else:
        
        print("Unfortunately! No Scholarship")
```

```python
std_db={
    'Student ID':student_ids,
    'Gender':gender,
    'Department':dept,
    'Sessional perc':sessional_perc,
    'Total Marks':total_marks,
    'Backlogs':backlogs,
    'Scholarships':scholarships
}
```
