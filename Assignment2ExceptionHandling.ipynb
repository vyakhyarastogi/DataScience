Using the custom exception, we can have your own exception and message. Here, we have passed a string to the constructor of superclass i.e. Exception class that can be obtained using getMessage() method on the object we have created.

```python
class Error(Exception):
    """ Base class for other exceptions"""
    print("This is a base class")
    pass

class zerodivision(Error):
    """Raised when the input value is zero"""
    print("Excuted"
         )
    pass

class print_value(Error):
    pass
    

try:
    num=int(input("enter a number:"))
    if num==0:
        raise zerodivision
    else:
        raise print_value
except zerodivision:
    print("Input value is zero,try again!")
except print_value:
    print(num)
    
```

```python
def floating_point_error_demo():
    try:
        # This will raise a ZeroDivisionError
        result = 1.0 / 0.0
        print("Result:", result)
    except ZeroDivisionError as e:
        print("Caught a ZeroDivisionError:", e)
```
